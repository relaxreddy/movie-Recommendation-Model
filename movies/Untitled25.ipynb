```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies_data=pd.read_csv("movies.csv")
```

```python
movies_data.head()
```

```python
# Number of rows and columns in a data frame
movies_data.shape
```

```python
# selecting the relevant features for recommendations
selected_features=['genres','title']
print(selected_features)
```

```python
# combining all selected features
combined_features = movies_data['genres']+' '+movies_data['title']
```

```python
print(combined_features)
```

```python
# converting the text data into feature vectors

vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

```python
# getting the similarity scores using cosine similarity

similarity = cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
print(similarity.shape)
```

```python
# getting movie names from user
movie_name = input("enter your favorite movie name : ")
```

```python
# creating a list with all movie names given in the dataset

list_of_all_titles = movies_data['title'].tolist()
print(list_of_all_titles)
```

```python
# finding the close match for the movie name given by the user
find_close_match = difflib.get_close_matches(movie_name,list_of_all_titles)
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```
